```python
Fuel.columns = [
    'FacilityId',
    'FRSId',
    'FacilityName',
    'City',
    'State',
    'PrimaryNAICSCode',
    'Year',
    'IndustryType',
    'Sector',
    'UnitName',
    'FuelType',
    'SpecificFuelType',
    'OtherFuelName',
    'BlendFuelName',
    'MethaneEmissions',
    'NitrousOxideEmissions',
]
Fuel.describe(include='all')
```

```python
FuelByCity = Fuel[pd.notna(Fuel.City)]
FuelByCity.City = FuelByCity.City.str.lower()
FuelByCity.sort_values('City')
```

```python
FuelByCity = FuelByCity.merge(income[['StateCode', 'City', 'StateAb', 'County', 'Lat', 'Lon', 'Median']], left_on=['City', 'State'], right_on=['City', 'StateAb'])
FuelByCity
```

```python
bins = np.arange(FuelByCity.Median.min(),
FuelByCity.Median.max()+1,
                 10000)

FuelByCity['IncomeGroups'] = FuelByCity.Median.apply(lambda x: bins[(bins - x < 10000) & (bins - x >= 0)][0])
# FuelByCity.IncomeGroups = FuelByCity.IncomeGroups.apply(lambda x: x.right)
FuelByCity.sort_values('IncomeGroups', ascending=False)
# pd.cut(FuelByCity.Median, bins, include_lowest=False)
```

```python
MethaneEmissions_ByIncome = FuelByCity[['Year','MethaneEmissions', 'IncomeGroups']].groupby(['Year', 'IncomeGroups']).sum('MethaneEmissions')
MethaneEmissions_ByIncome.reset_index(inplace=True)
MethaneEmissions_ByIncome
```

```python
alt.Chart(MethaneEmissions_ByIncome).mark_bar().encode(
    x='IncomeGroups:O',
    y='MethaneEmissions:Q',
    color = 'MethaneEmissions:Q'
).facet(
    facet='Year:O',
    columns=5
)
```

Emissions Over Time by Sector or Industry

```python
TopTenSectors = np.array(Fuel[['Sector']].value_counts()[:10].reset_index()['Sector'])
TopTenIndustries = np.array(Fuel[['IndustryType']].value_counts()[:10].reset_index()['IndustryType'])
```

```python
MethaneEmissionsBySector = Fuel[(Fuel.Sector.isin(TopTenSectors)) & (Fuel.Year > 2010)][['Year', 'Sector', 'MethaneEmissions']].groupby(['Year', 'Sector']).mean('MethaneEmissions')
MethaneEmissionsBySector.reset_index(inplace=True)
```

```python
MethaneEmissions_BySector = alt.Chart(MethaneEmissionsBySector).mark_line().encode(
    x = 'Year:O',
    y = 'MethaneEmissions:Q',
).facet(
    facet='Sector:N',
    columns = 5
)
```

Emission Types in Unit

```python
Unit.columns = [
    'FacilityId',
    'FRSId',
    'FacilityName',
    'City',
    'State',
    'PrimaryNAICSCode',
    'Year',
    'IndustryType',
    'Sector',
    'UnitName',
    'UnitType',
    'Method',
    'MaxHeat',
    'CO2EmissionsNonBiogenic',
    'MethaneEmissions',
    'NitrousOxideEmissions',
    'CO2EmissionsBiogenic'
]
```

```python
Unit.describe(include='all')
```

```python
UnitEmissions = Unit[(Unit.Year > 2010)][[
    'FacilityId',
    'Year',
    'IndustryType',
    'Sector',
    'CO2EmissionsNonBiogenic',
    'MethaneEmissions',
    'NitrousOxideEmissions',
    'CO2EmissionsBiogenic'
    ]]
UnitEmissions = pd.melt(
    Unit,id_vars=[
        'FacilityId',
        'Year',
        'IndustryType',
        'Sector'
        ],
    value_vars=[
        'CO2EmissionsNonBiogenic',
        'MethaneEmissions',
        'NitrousOxideEmissions',
        'CO2EmissionsBiogenic'
        ],
    var_name='EmissionType',
    value_name='Emissions')

UnitEmissionsByYear = UnitEmissions.groupby(['Year','EmissionType']).mean('Emissions').reset_index()
```

```python
UnitEmissionsByYear
```

```python
alt.Chart(UnitEmissionsByYear).mark_line().encode(
    x = 'Year:O',
    y = 'Emissions:Q',
    color = 'EmissionType:N'
)
```

```python
alt.Chart(UnitEmissionsByYear).mark_bar().encode(
    x = alt.X('sum(Emissions)', stack = 'normalize'),
    y = 'Year:O',
    color = 'EmissionType:N'
)
```
